```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import math as m

import matplotlib.pyplot as plt
import numpy as np

def bohachevsky_1(x):
    return x[0]*x[0] + 2.0*x[1]*x[1]  \
    - 0.3*np.cos(3.0*np.pi*x[0]) - 0.4*np.cos(4.0*np.pi*x[1]) \
    + 0.7
    
def bohachevsky_2(x):
    return x[0]*x[0] + 2.0*x[1]*x[1]  \
    - 0.3*np.cos(3.0*np.pi*x[0])*np.cos(4.0*np.pi*x[1]) \
    + 0.3
    
def bohachevsky_3(x):
    return x[0]*x[0] + 2.0*x[1]*x[1]  \
    - 0.3*np.cos(3.0*np.pi*x[0] + 4.0*np.pi*x[1]) \
    + 0.3
    
def grad_1(x):
    zx =2*x[0] + 3.0*np.pi*0.3*np.sin(3.0*np.pi*x[0])
    zy =4*x[1] + 4.0*np.pi*0.4*np.sin(4.0*np.pi*x[1])
    return np.array(zx,zy)

def grad_2(x):
    zx =2*x[0] + 3.0*np.pi*0.3*np.sin(3.0*np.pi*x[0])*np.cos(4.0*np.pi*x[1])
    zy =4*x[1] + 4.0*np.pi*0.4*np.sin(4.0*np.pi*x[1])*np.cos(3.0*np.pi*x[0])
    return np.array(zx,zy)

def grad_3(x):
    zx =2*x[0] + 3.0*np.pi*0.3*np.sin(3.0*np.pi*x[0] + 4.0*np.pi*x[1])
    zy =4*x[1] + 4.0*np.pi*0.4*np.sin(3.0*np.pi*x[0] + 4.0*np.pi*x[1])
    return np.array(zx,zy)

fig = plt.figure()
ax = Axes3D(fig, azim = -29, elev = 50)
# ax = Axes3D(fig)
X = np.arange(-100, 100, 1)
Y = np.arange(-100, 100, 1)

X, Y = np.meshgrid(X, Y)
Z = np.fromiter(map(bohachevsky_3, zip(X.flat,Y.flat)), dtype=np.float, count=X.shape[0]*X.shape[1]).reshape(X.shape)


ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet, linewidth=0.2)
 
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
def three_hump_camel(x):
    term1 = 2*x[0]**2
    term2 = -1.05*x[0]**4
    term3 = x[0]**6 / 6
    term4 = x[0]*x[1]
    term5 = x[1]**2
    return term1 + term2 + term3 + term4 + term5

def grad_three_hump_camel(x):
    zx =4*x[0] - 4.0*1.05*x[0]**3 + x[0]**5 + x[1]
    zy =2*x[1] + x[0]
    return np.array(zx,zy)

fig = plt.figure()
ax = Axes3D(fig, azim = -29, elev = 50)
# ax = Axes3D(fig)
X = np.arange(-2, 2, 0.05)
Y = np.arange(-2, 2, 0.05)
X, Y = np.meshgrid(X, Y)
Z = np.fromiter(map(three_hump_camel, zip(X.flat,Y.flat)), dtype=np.float, count=X.shape[0]*X.shape[1]).reshape(X.shape)


ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet, linewidth=0.2)
 
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
def rosenbrock(x):
    return 100*(x[1]-x[0]**2)**2 + (x[0]-1)**2

def grad_rosenbrock(x):
    zx = -400*(x[1]-x[0]**2)*x[0] + 2*(x[0]-1)
    zy = 200*(x[1]-x[0]**2)
    return np.array(zx,zy)


fig = plt.figure()
ax = Axes3D(fig, azim = -29, elev = 50)
# ax = Axes3D(fig)
X = np.arange(-10, 10, 0.5)
Y = np.arange(-6, 6, 0.5)
X, Y = np.meshgrid(X, Y)
Z = np.fromiter(map(rosenbrock, zip(X.flat,Y.flat)), dtype=np.float, count=X.shape[0]*X.shape[1]).reshape(X.shape)


ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet, linewidth=0.2)
 
plt.xlabel("x")
plt.ylabel("y")

plt.show()

```

```python
def beale(x):
    x1 = x[0]
    x2 = x[1]
    term1 = (1.5 - x1 + x1*x2)**2
    term2 = (2.25 - x1 + x1*x2**2)**2
    term3 = (2.625 - x1 + x1*x2**3)**2
    return term1 + term2 + term3

def grad_beale(x):
    x1 = x[0]
    x2 = x[1]
    zx = 2*(1.5 - x1 + x1*x2)*(-1+x2) + 2*(2.25 - x1 + x1*x2**2)*(-1+x2**2) + 2*(2.625 - x1 + x1*x2**3)*(-1+x2**3)
    zy = 2*(1.5 - x1 + x1*x2)*(x1) + 2*(2.25 - x1 + x1*x2**2)*(x1*2*x2) + 2*(2.625 - x1 + x1*x2**3)*(x1*3*x2**2)
    return np.array(zx,zy)
    

fig = plt.figure()
ax = Axes3D(fig, azim = -29, elev = 50)
# ax = Axes3D(fig)
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
Z = np.fromiter(map(beale, zip(X.flat,Y.flat)), dtype=np.float, count=X.shape[0]*X.shape[1]).reshape(X.shape)


ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet, linewidth=0.2)
 
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
def styblinsky_tang(x):
    x1 = x[0]
    x2 = x[1]
    return x1**4 - 16*x1**2 + 5*x1 + x2**4 - 16*x2**2 + 5*x2

def grad_styblinsky_tang(x):
    x1 = x[0]
    x2 = x[1]
    zx = 4*x1**3 - 32*x1 + 5
    zy = 4*x2**3 - 32*x2 + 5
    return np.array(zx,zy)

fig = plt.figure()
ax = Axes3D(fig, azim = -29, elev = 50)
# ax = Axes3D(fig)
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
Z = np.fromiter(map(styblinsky_tang , zip(X.flat,Y.flat)), dtype=np.float, count=X.shape[0]*X.shape[1]).reshape(X.shape)


ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet, linewidth=0.2)
 
plt.xlabel("x")
plt.ylabel("y")

plt.show()

```

```python
def mcCormick(x):
    x1 = x[0]
    x2 = x[1]
    term1 = np.sin(x1 + x2)
    term2 = (x1 - x2)**2
    term3 = -1.5*x1
    term4 = 2.5*x2
    return term1 + term2 + term3 + term4 + 1

def grad_McCormick(x):
    x1 = x[0]
    x2 = x[1]
    zx = -np.cos(x1 + x2) + 2*(x1 - x2) - 1.5
    zy = -np.cos(x1 + x2) - 2*(x1 - x2) + 2.5
    return np.array(zx,zy)
    

fig = plt.figure()
ax = Axes3D(fig, azim = -29, elev = 50)
# ax = Axes3D(fig)
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X, Y = np.meshgrid(X, Y)
Z = np.fromiter(map(mcCormick , zip(X.flat,Y.flat)), dtype=np.float, count=X.shape[0]*X.shape[1]).reshape(X.shape)


ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet, linewidth=0.2)
 
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
